```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 범주형 데이터 시각화
```

```python
# 예시로 사용할 분할표 데이터
data = [[5,10,15],[10,15,20]]

# seaborn을 이용한 히트맵 시각화
sns.heatmap(data, cmap='rainbow', annot=True, fmt='d', cbar=False)

# 플롯 제목 추가
plt.title('Example Contigency Table Heatmap')

#플롯 출력
plt.show()
```
