```python
import pandas as pd
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 학년별 데이터 생성
first_df = pd.DataFrame({'Grade': 1}, index=pd.RangeIndex(100), columns=['Grade'])
second_df = pd.DataFrame({'Grade': 2}, index=pd.RangeIndex(100), columns=['Grade'])
third_df = pd.DataFrame({'Grade': 3}, index=pd.RangeIndex(100), columns=['Grade'])

# 각 학년의 과목 점수 열 추가
subjects = ['Math', 'English', 'Korean']
for subject in subjects:
    first_df[subject] = np.random.normal(loc=70, scale=10, size=(100,))
    second_df[subject] =np.random.normal(loc=80, scale=12, size=(100,))
    third_df[subject] = np.random.normal(loc=60, scale=8, size=(100,))

print(first_df)
print(second_df)
print(third_df)
```

```python
# 데이터 프레임을 모두 합쳐 하나의 데이터 프레임으로 만들기 
df = pd.concat([first_df,second_df,third_df])
print(df)
```

```python
# csv 파일로 저장 
df.to_csv("./data/student_scores.csv")
```

```python
# 상관계수 계산 
corr = df.corr()

print(corr)
```

```python
# 히트맵 출력
sns.heatmap(corr, annot=True, cmap='plasma')
plt.show()
```

```python
math_stats = df['Math'].describe()
english_stats = df['English'].describe()
korean_stats = df['Korean'].describe()

print(math_stats)
print(english_stats)
print(korean_stats)
```

```python
# 수학 성적 분포 시각화
plt.hist(df['Math'], bins=10)
plt.title('Math Scores Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()

# 영어 성적 분포 시각화 
plt.hist(df['English'], bins=10)
plt.title('English Scores Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()

# 국어 성적 분포 시각화
plt.hist(df['Korean'], bins=10)
plt.title('Korean Scores Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```
