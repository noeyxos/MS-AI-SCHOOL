```python
import numpy as np
import pandas as pd
from scipy.stats import mannwhitneyu
```

```python
# 랜덤 데이터 생성
np.random.seed(42)
group1 = np.random.normal(loc=6,scale=1.5,size=100)
group2= np.random.normal(loc=7,scale=1.5,size=100)

#csv 저장
data = pd.DataFrame({
    'group1': group1,
    'group2': group2})


data.to_csv('./data/sleep_data.csv', index=False, encoding='utf-8')
```

```python
# csv 읽기
sleep_data = pd.read_csv('./data/sleep_data.csv')
print(sleep_data)
```

```python
group1_data = data['group1']
group2_data = data['group2']

print(group1)
print(group2)
```

```python
# 앤 - 휘트니 검정 수행
st, p_val = mannwhitneyu(group1_data,group2_data, alternative='two-sided') # 양측검증
print('Mann-Whitney U statistic : ',st)
print('p_value: ',p_val)
```

```python
# 유의수준 0.05로 가설 검정
if p_val < 0.05:
    print('두 집단 간에는 유의한 차이가 있습니다.')
else:
    print('두 집단 간에는 유의한 차이가 없습니다.')
```
