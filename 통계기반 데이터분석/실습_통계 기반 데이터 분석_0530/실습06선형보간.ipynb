```python
# 선형보간 - pandas를 활용한 선형 보간 실습
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 시계열 데이터 생성 
dates = pd.date_range(start='2023-01-01', periods=10, freq='D')
values = [1, np.nan,3, np.nan, 5, 6, np.nan, 8, 9,np.nan,]
```

```python
# 데이터 프레임 생성
data = pd.DataFrame({'Date': dates, 'Values':values})
print(data)
```

```python
# 선형 보간 적용
data['Values_linear'] = data['Values'].interpolate(method='linear')
print(data)

# 시계열 데이터 시각화
plt.plot(data['Date'], data['Values'], 'o-', label='Original')
plt.plot(data['Date'], data['Values_linear'], 'x-', label='Linear Interpolation')
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Linear Interpolation')
plt.xticks(rotation=-45)
plt.show()
```
