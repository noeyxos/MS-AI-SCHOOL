### 상향, 하향, 평평한 추세 시각화

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 랜덤 시계열 데이터 생성 총 100개
np.random.seed(42)
dates = pd.date_range(start='2023-01-01',periods=100, freq='D')
values = np.random.randn(100).cumsum() # cumsum(): 누적합

# 데이터 프레임 생성
data = pd.DataFrame({'Date':dates, 'Value':values})

print(data)
```

```python
# 상향 추세 시각화 
data['Upward_Trend'] = np.linspace(0,50,len(data)) # 상향 추세 생성
print(data)
plt.plot(data['Date'], data['Upward_Trend'], label = ['Upward_Trend'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Upward_Trend')
plt.legend()
plt.xticks(rotation=-45)
plt.show()
```

```python
# 하향 추세 시각화 
data['Downward_Trend'] = np.linspace(50,0,len(data)) # 하향 추세 생성
print(data)
plt.plot(data['Date'], data['Downward_Trend'], label = ['Downward_Trend'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Downward_Trend')
plt.legend()
plt.xticks(rotation=-45)
plt.show()
```

```python
# 평평한 추세 시각화 
data['Flat_Trend'] = np.full(len(data),50) # 하향 추세 생성
print(data)
plt.plot(data['Date'], data['Flat_Trend'], label = ['Flat_Trend'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Flat_Trend')
plt.legend()
plt.xticks(rotation=-45)
plt.show()
```
