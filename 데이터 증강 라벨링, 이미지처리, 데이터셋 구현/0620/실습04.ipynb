# 원본 오디오와 잡음 생성된 오디오 비교

```python
import soundfile as sf
import matplotlib.pyplot as plt

# 오디오 듣기 위해서 사용할 라이브러리 
import IPython.display as ipd
# ipd.Audio(y, rate=sr)
```

### 원본 오디오 불러오기

```python
file_path = './blues.00001.wav' 
# wav: 일반적으로 windows에서 사용되는 오디오 파일 형식 -> 고해상도 및 대용량 오디오 파일 저장하기 위해서 사용
data, sr = sf.read(file_path)
```

### 잡음 오디오 불러오기

```python
file_path = './sample_noise_added.wav' 
# wav: 일반적으로 windows에서 사용되는 오디오 파일 형식 -> 고해상도 및 대용량 오디오 파일 저장하기 위해서 사용
noise_data, noise_sr = sf.read(file_path)
```

### 시각화 

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.plot(data)
plt.title('Original')
plt.subplot(1,2,2)
plt.plot(noise_data)
plt.title('Noise Added')
plt.show()
```
