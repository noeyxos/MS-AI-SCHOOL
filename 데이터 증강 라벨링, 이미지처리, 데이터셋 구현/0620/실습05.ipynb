```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import soundfile as sf
```

### 1. 음성 데이터 불러오기 

```python
data, sr = librosa.load('./blues.00001.wav')
```

### 2. stretch the audio signal

```python
data_stretch = librosa.effects.time_stretch(data, rate=0.8)
```

### 3. 원본 데이터와 늘린 데이터 비교

```python
plt.figure(figsize=(14,5))
```

### 4. 결과 저장

```python
save_path = './sample_stretch.wav'
sf.write(save_path, data_stretch, sr)
```

```python
plt.subplot(1,2,1)
librosa.display.waveshow(data, sr=sr, alpha=0.8)
plt.title('Original')

plt.subplot(1,2,2)
librosa.display.waveshow(data_stretch, sr=sr, alpha=0.8)

plt.title('Stretch')

plt.tight_layout()
plt.show()
```
