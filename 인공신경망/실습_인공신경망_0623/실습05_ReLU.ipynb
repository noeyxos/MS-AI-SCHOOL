# 인공신경망 - ReLU(렐루) 함수 실습

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def relu(x):
    return np.maximum(0, x)

x = np.linspace(-5, 5, 1000)
y = relu(x)

print(relu(-5)) # 0
print(relu(5)) # 0
print(relu(1000)) # 5

plt.plot(x, y)
plt.title('ReLU Function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Leaky ReLU 함수

```python
import matplotlib.pyplot as plt
import numpy as np

def leaky_relu(x, alpha = 0.01):
    return np.where(x >= 0, x, x * alpha)

x = np.linspace(-5, 5, 100)
y = leaky_relu(x)

plt.plot(x,y)
plt.title('Leaky ReLU Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()
```
