# 인공신경망 - 소프트맥스

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def softmax(z):
    return np.exp(z) / np.sum(np.exp(z))


z = np.linspace(-10, 10, 100 ) # -10부터 10까지 100개의 값을 생성
y = softmax(z)

plt.plot(z, y)
plt.title("Softmax Function")
plt.xlabel("z")
plt.ylabel("softmax(z)")
plt.show()
```
