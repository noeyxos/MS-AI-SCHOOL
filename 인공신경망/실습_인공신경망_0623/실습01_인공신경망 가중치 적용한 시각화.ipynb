# 인공신경망 가중치 적용한 시각화 실습

```python
import torch
import matplotlib.pyplot as plt
```

```python
# 2차원 가중치 행렬을 랜덤으로 초기화
w = torch.randn(3, 3)
print(w)


# 가중치를 1차원 벡터로 변환
w_flattened = w.flatten()

# 가중치 분포를 시각화
plt.hist(w_flattened, bins=20)
plt.xlabel('Weight Value')
plt.ylabel('Frequency')
plt.show()
```

### 크기가 1인 랜덤한 값을 가지는 편향을 생성

```python
# 편향을 랜덤으로 초기화
b = torch.randn(1)

# 편향 값을 시각화 
plt.hist(b.detach().numpy(), bins=20) 
#b.detach().numpy() : 텐서를 넘파이 배열로 변환

plt.xlabel('Bias Value')
plt.ylabel('Frequency')
plt.show()
```
