# 인공신경망 - 하이퍼볼릭 탄젠트 함수 실습

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tanh(x):
    return np.tanh(x)


x = np.linspace(-5, 5,100)
y = tanh(x)

print(tanh(-5))
print(tanh(5))
print(tanh(100))


plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Hyperbolic Tangent Function')
plt.show()
```
