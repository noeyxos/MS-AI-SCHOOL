# 인공지능 신경망 - 활성화 함수 시그모이드 실습

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100) # -10부터 10까지 100의 값을 생성합니다.
y = sigmoid(x)

print(sigmoid(-10))
print(sigmoid(10))
print(sigmoid(0))
print(sigmoid(100))

plt.plot(x,y)
plt.title('Sigmoid Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()
```
