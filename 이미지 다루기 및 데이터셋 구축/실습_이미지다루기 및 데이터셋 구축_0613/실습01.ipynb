```python
import cv2
import matplotlib.pyplot as plt
```

```python
# 이미지 읽기
image =cv2.imread('image01.jpeg')

# BGR -> RGB 변경
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 이미지 밝기 조정
gamma = 2.5
img_corrected = cv2.pow(image / 255.0 , gamma)
img_corrected = img_corrected * 255.0

# 원본 이미지 밝기 조정 결과 모습
plt.imshow(image)
plt.title("org image")
plt.show()

# 이미지 밝기 조정 결과 모습
plt.imshow(img_corrected)
plt.title("img_corrected")
plt.show()
```
